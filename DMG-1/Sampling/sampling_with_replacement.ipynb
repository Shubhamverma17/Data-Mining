## SAMPLING WITH REPLACEMENT

```python
import random
import matplotlib.pyplot as plt
import collections
num=[]
for i in range(1000):
    num.append(random.randint(0,999))
```

```python
def sample_data(k):
    test=[]
    loop=True
    flag={i:0 for i in range(10)}
    while(loop):
        temp=num[random.randint(0,999)]
        test.append(temp)
        position=(temp//100)
        flag[position]+=1
        i=0
        valid=True
        while(valid and i<10):
            if flag[i]<k:
                valid=False
            i+=1
        if valid==True:
            loop=False
    return len(test)
```

```python
def analysis(k):
    sample_info=[]
    for i in range(1000):
        sample_info.append(sample_data(k))
    sample_info.sort()
    print(f"minimum sample size- {min(sample_info)}, maximum sample size- {max(sample_info)}, total sample- {len(sample_info)}")

    count=collections.Counter(sample_info)
    keys=list(count.keys())
    freq=list(count.values())
    c_freq=[]
    sum=0
    for c in freq:
        sum+=c
        c_freq.append(sum/1000)
    
    plt.bar(keys,freq,width =1, color = ['red', 'green','blue','black'])
    plt.xlabel('Sample Size') 
    plt.ylabel('Frequency') 
    plt.title(f'Sample Size - Frequency Graph K={k}')  
    plt.show() 

    plt.hist(sample_info,edgecolor='red', bins=10)
    plt.xlabel('Sample Size range') 
    plt.ylabel('Frequency') 
    plt.title(f'Sample Size range(10) - Frequency Graph K={k}')  
    plt.show()
    
    plt.plot(keys,c_freq,marker='.',linestyle='none')
    plt.xlabel('Sample Size') 
    plt.ylabel('Cumulative Probability Graph') 
    plt.title(f'Sample Size Vs Cumulative Probability Graph K={k}')  
    
```

THERE ARE 3 GRAPH FOR K=1,2,3,4
GRAPH 1: SAMPLE SIZE VS FREQUENCY OF EACH SAMPLE SIZE
GRAPH 2: SAMPLE SIZE INTERVAL OF 10 IS CREATED THEN (INTERVAL VS FREQUENCY GRAPH HAS BEEN PLOTTED)
GRAPH 3: SAMPLE SIZE VS CUMULATIVE PROBABILITY

```python
analysis(1)
#k=1
```

```python
analysis(2)
#k=2
```

```python
analysis(3)
#k=3
```

```python
analysis(4)
#k=4
```

## CONCLUSION: INCREASING THE VALUE OF K, SAMPLE SIZE WILL ALSO INCREASE (KNOWING THE FACT THAT MORE NUMBER OF ENTRIES WILL BE SAMPLED)
